```python
import pandas as pd
from sklearn.preprocessing  import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv('Tourism.csv')
df.nunique()
```

## Encode categorical Variables

```python
df=pd.read_csv('Tourism.csv')
df['CityName_Frequency']=df['CityName'].map(df['CityName'].value_counts())
df['CityId_Frequency']=df['CityId'].map(df['CityId'].value_counts())
df['Country_Frequency']=df['Country'].map(df['Country'].value_counts())
df['Region_Frequency']=df['Region'].map(df['Region'].value_counts())
df['Attraction_Frequency']=df['Attraction'].map(df['Attraction'].value_counts())
df['AttractionAddress_Frequency']=df['AttractionAddress'].map(df['AttractionAddress'].value_counts())
df['AttractionType_Frequency']=df['AttractionType'].map(df['AttractionType'].value_counts())
le=LabelEncoder()
Column_to_Label_Encoder=['Continent','VisitMode_y']
for col in Column_to_Label_Encoder:
    df[f'{col}_LB_EC']=le.fit_transform(df[col])

df.to_csv('Tourism.csv',index=False)
```

```python
df.nunique()
```

## Aggregate

```python
User_Profile = df.groupby('UserId').agg(
    Avg_Rating_Per_Visit=('Rating','mean'),
    ToTal_Visits=('VisitMode_x','count'),
    Avg_Spending=('TransactionId','mean'),
    Distinct_Attraction=('AttractionId','nunique'),
    Preferred_Visit_Mode=("VisitMode_y",lambda x : x.mode()[0] if not x.mode().empty else None )
).reset_index()

User_Profile.head(5)
```
