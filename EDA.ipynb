```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df =  pd.read_csv('Tourism.csv')
```

```python
plt.figure(figsize=(12,5))
sns.countplot(data=df,x='Continent',order= df['Continent'].value_counts().index,palette='cividis')
plt.xlabel('Traveler Visited Continent')
plt.ylabel('No.Of Traveler Visited')
plt.title("Distrubution Of Continent Traveled")
plt.show()
```

```python
plt.figure(figsize=(12,5))
top_countries=df['Country'].value_counts().nlargest(10)
sns.barplot( x=top_countries.index,y=top_countries.values ,palette="coolwarm")
plt.xlabel('Traveler Visited Country')
plt.ylabel('No Of Traveler Visited')
plt.title('Distrubution OF Country Traveled')
plt.show()
```

```python
plt.figure(figsize=(12,5))
Top_Region=df['Region'].value_counts().nlargest(10)
sns.barplot( x=Top_Region.index,y=Top_Region.values ,palette="magma")
plt.xlabel('Traveler Visited Region')
plt.ylabel('No Of Traveler Visited')
plt.title('Distrubution OF Region Traveled')
plt.show()
```

```python
attraction_ratings = df.groupby("AttractionType")["Rating"].mean().reset_index()

attraction_ratings = attraction_ratings.sort_values(by="Rating", ascending=False)

plt.figure(figsize=(14, 6))
sns.barplot(data=attraction_ratings, x="Rating", y="AttractionType", palette="viridis")
plt.xlabel("Average User Rating")
plt.ylabel("Attraction Type")
plt.title("Popularity of Attraction Types Based on User Ratings")
plt.show()
```

```python
numerical_df = df.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numerical_df.corr()
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', center=0, linewidths=0.5)
plt.title('Correlation Matrix of Numerical Features', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
visit_mode_demo = df[["VisitMode_y_LB_EC", "Continent_LB_EC", "Country_Frequency", "Region_Frequency"]]

visit_mode_distribution = visit_mode_demo.groupby(["VisitMode_y_LB_EC", "Continent_LB_EC"]).size().unstack()

plt.figure(figsize=(10, 6))
sns.heatmap(visit_mode_distribution, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Visit Mode Distribution Across Continents")
plt.xlabel("Continent")
plt.ylabel("Visit Mode")
plt.show()
```

```python
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.boxplot(data=df, x="Rating", y="Attraction", palette="magma")
plt.xlabel("Rating")
plt.ylabel("Attraction")
plt.title("Distribution of Ratings Across Attractions")

plt.subplot(1, 2, 2)
sns.boxplot(data=df, x="Rating", y="Region", palette="coolwarm")
plt.xlabel("Rating")
plt.ylabel("Region")
plt.title("Distribution of Ratings Across Regions")
plt.tight_layout()
plt.show()
```
